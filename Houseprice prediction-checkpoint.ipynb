```python
#Model preparation
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
Y=train["SalePrice"].values
X=train[features_].values
test=test[features_].values
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
#Linear regression model
ll=LinearRegression()
ll.fit(X_train,Y_train)
Y_pred_test=ll.predict(X_test)
Y_pred_train=ll.predict(X_train)
```

```python
#model evaluation
mse_test=mean_squared_error(Y_test,Y_pred_test)
mse_train=mean_squared_error(Y_train,Y_pred_train)
r2_test=r2_score(Y_test,Y_pred_test)
r2_train=r2_score(Y_train,Y_pred_train)
print("Test set:"+ "  MSE="+str(mse_test)+ "  R2="+str(r2_test))
print("Train set:"+ "  MSE="+str(mse_train)+ "  R2="+str(r2_train))
```

```python
plt.plot(Y_test,label="Test set")
plt.plot(Y_pred_test,label="Predictions")
```

```python
from sklearn.linear_model import Ridge,Lasso,ElasticNet
```

```python
#Regularization L2
alphas=[0.0001,0.001,0.01,0.1,1,10]
for i in alphas:
    model=Ridge(alpha=i)
    model.fit(X_train,Y_train)
    Y_pred_test=model.predict(X_test)
    Y_pred_train=model.predict(X_train)
    mse_test=mean_squared_error(Y_test,Y_pred_test) 
    mse_train=mean_squared_error(Y_train,Y_pred_train)
    r2_test=r2_score(Y_test,Y_pred_test)
    r2_train=r2_score(Y_train,Y_pred_train)
    print("ALPHA"+ str(i))
    print("Test set:"+ "  MSE="+str(mse_test)+"  R2="+str(r2_test))
    print("Train set:"+ "  MSE="+str(mse_train)+"  R2="+str(r2_train))
```

```python
#Regularization l2
alphas=[0.0001,0.001,0.01,0.1,1,10]
for i in alphas:
    model=Lasso(alpha=i)
    model.fit(X_train,Y_train)
    Y_pred_test=model.predict(X_test)
    Y_pred_train=model.predict(X_train)
    mse_test=mean_squared_error(Y_test,Y_pred_test) 
    mse_train=mean_squared_error(Y_train,Y_pred_train)
    r2_test=r2_score(Y_test,Y_pred_test)
    r2_train=r2_score(Y_train,Y_pred_train)
    print("ALPHA"+ str(i))
    print("Test set:"+ "  MSE="+str(mse_test)+"  R2="+str(r2_test))
    print("Train set:"+ "  MSE="+str(mse_train)+"  R2="+str(r2_train))

```

```python
#Regularization l2 and l1
alphas=[0.0001,0.001,0.01,0.1,1,10]
for i in alphas:
    model=ElasticNet(alpha=i,l1_ratio=0.5)
    model.fit(X_train,Y_train)
    Y_pred_test=model.predict(X_test)
    Y_pred_train=model.predict(X_train)
    mse_test=mean_squared_error(Y_test,Y_pred_test) 
    mse_train=mean_squared_error(Y_train,Y_pred_train)
    r2_test=r2_score(Y_test,Y_pred_test)
    r2_train=r2_score(Y_train,Y_pred_train)
    print("ALPHA"+ str(i))
    print("Test set:"+ "  MSE="+str(mse_test)+"  R2="+str(r2_test))
    print("Train set:"+ "  MSE="+str(mse_train)+"  R2="+str(r2_train))
```

```python
#Random forest regressor model
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor(n_estimators=30)
rfr.fit(X_train,Y_train)
```

```python
rfr_test=rfr.predict(X_test)
rfr_train=rfr.predict(X_train)
```

```python
#Model evaluation
mse_test=mean_squared_error(Y_test,rfr_test)
r2_test=r2_score(Y_test,rfr_test)
mse_train=mean_squared_error(Y_train,rfr_train)
r2_train=r2_score(Y_train,rfr_train)
print("Test set"+"  MSE="+str(mse_test)+"  R2="+str(r2_test))
print("Train set:"+ "  MSE="+str(mse_train)+"  R2="+str(r2_train))
```
